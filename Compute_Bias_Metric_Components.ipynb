CALCULATION

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_distances

# === Load input CSVs ===
frame_df = pd.read_csv("/content/drive/MyDrive/FINAL DATASET/frame_level_feature_data.csv")
embed_df = pd.read_csv("/content/drive/MyDrive/FINAL DATASET/video_level_embedding_data.csv")
gender_df = pd.read_csv("/content/drive/MyDrive/MAY19/video_metadata_with_gender.csv")

# === Size Bias ===
frame_df["size_ratio"] = frame_df["bbox_area"] / frame_df["frame_area"]
size_bias_df = frame_df.groupby("video_name")["size_ratio"].mean().reset_index(name="size_bias")
size_bias_df = size_bias_df.merge(gender_df[["video_name", "gender"]], on="video_name")
size_bias_df["size_bias"] *= size_bias_df["gender"].map({"male": 1, "female": -1})
size_bias_df = size_bias_df.drop(columns=["gender"])

# === Centering Bias ===
frame_df["frame_diag_half"] = 0.5 * np.sqrt(frame_df["frame_width"]**2 + frame_df["frame_height"]**2)
frame_df["center_norm"] = 1 - (frame_df["distance_to_center"] / frame_df["frame_diag_half"])
centering_bias_df = frame_df.groupby("video_name")["center_norm"].mean().reset_index(name="centering_bias")
centering_bias_df = centering_bias_df.merge(gender_df[["video_name", "gender"]], on="video_name")
centering_bias_df["centering_bias"] *= centering_bias_df["gender"].map({"male": 1, "female": -1})
centering_bias_df = centering_bias_df.drop(columns=["gender"])

# === Screen Time Bias ===
screen_time_df = frame_df.groupby("video_name").size().reset_index(name="detected_frames")
screen_time_df = screen_time_df.merge(gender_df[["video_name", "gender"]], on="video_name")

# Apply sign based on gender (+1 for male, -1 for female)
screen_time_df["screen_time_bias"] = screen_time_df["detected_frames"] * screen_time_df["gender"].map({"male": 1, "female": -1})

# Normalize by total detected frames for interpretability
total_frames = screen_time_df["detected_frames"].sum()
screen_time_df["screen_time_bias"] = screen_time_df["screen_time_bias"] / total_frames

screen_time_bias_df = screen_time_df[["video_name", "screen_time_bias"]]


# === Embedding Bias (with L2 normalization) ===
embed_full = embed_df.merge(gender_df[["video_name", "gender"]], on="video_name")
embed_cols = [c for c in embed_df.columns if c.startswith("embed_")]

def normalize_vector(v):
    norm = np.linalg.norm(v)
    return v / norm if norm > 0 else v

embeddings = embed_full[embed_cols].values
norms = np.linalg.norm(embeddings, axis=1, keepdims=True)
norms[norms == 0] = 1
embed_full[embed_cols] = embeddings / norms
embed_full = embed_full.copy()

male_vectors = embed_full[embed_full["gender"] == "male"][embed_cols]
female_vectors = embed_full[embed_full["gender"] == "female"][embed_cols]

male_centroid = normalize_vector(male_vectors.mean().values).reshape(1, -1) if not male_vectors.empty else None
female_centroid = normalize_vector(female_vectors.mean().values).reshape(1, -1) if not female_vectors.empty else None

def embedding_bias(row):
    vec = row[embed_cols].values.reshape(1, -1)
    dm = cosine_distances(vec, male_centroid)[0][0] if male_centroid is not None else None
    df = cosine_distances(vec, female_centroid)[0][0] if female_centroid is not None else None
    if dm is not None and df is not None:
        return df - dm
    elif dm is not None:
        return dm - 0.5
    elif df is not None:
        return 0.5 - df
    else:
        print(f"Skipping embedding bias for video (missing centroids): {row['video_name']}")
        return 0

embed_full["embedding_bias"] = embed_full.apply(embedding_bias, axis=1)
embedding_bias_df = embed_full[["video_name", "embedding_bias"]]

# === Motion Bias (normalize keypoints) ===
pose_cols = [c for c in frame_df.columns if c.startswith("pose_")]
motion_vectors = []
skipped_motion = []

for video, group in frame_df.groupby("video_name"):
    fw = group.iloc[0]["frame_width"]
    fh = group.iloc[0]["frame_height"]
    try:
        poses = group.sort_values("frame")[pose_cols].dropna().values.astype(float).reshape(-1, 2, int(len(pose_cols)/2))
        poses[:, 0, :] /= fw
        poses[:, 1, :] /= fh
        flat_norm_poses = poses.reshape(poses.shape[0], -1)
        if len(flat_norm_poses) < 2:
            skipped_motion.append((video, "<2 pose frames"))
            continue
        motion = np.abs(np.diff(flat_norm_poses, axis=0)).mean(axis=0)
        motion_vectors.append((video, motion))
    except Exception as e:
        skipped_motion.append((video, f"reshape error: {e}"))

for vid, reason in skipped_motion:
    print(f"Skipping motion bias for {vid}: {reason}")

motion_df = pd.DataFrame(motion_vectors, columns=["video_name", "motion_vector"])
motion_df = motion_df.merge(gender_df[["video_name", "gender"]], on="video_name")

male_motions = [v for v, g in zip(motion_df["motion_vector"], motion_df["gender"]) if g == "male"]
female_motions = [v for v, g in zip(motion_df["motion_vector"], motion_df["gender"]) if g == "female"]

male_motion_centroid = np.mean(male_motions, axis=0) if male_motions else None
female_motion_centroid = np.mean(female_motions, axis=0) if female_motions else None

def motion_bias(vec):
    dm = cosine_distances([vec], [male_motion_centroid])[0][0] if male_motion_centroid is not None else None
    df = cosine_distances([vec], [female_motion_centroid])[0][0] if female_motion_centroid is not None else None
    if dm is not None and df is not None:
        return df - dm
    elif dm is not None:
        return dm - 0.5
    elif df is not None:
        return 0.5 - df
    else:
        return 0

motion_df["motion_bias"] = motion_df["motion_vector"].apply(motion_bias)
motion_bias_df = motion_df[["video_name", "motion_bias"]]

# === Final Merge ===
bias_df = size_bias_df \
    .merge(centering_bias_df, on="video_name") \
    .merge(screen_time_bias_df, on="video_name") \
    .merge(embedding_bias_df, on="video_name") \
    .merge(motion_bias_df, on="video_name")

# Save or display
bias_df.to_csv("/content/drive/MyDrive/MAY19/final_bias_metrics_values.csv", index=False)
print("Final bias metrics saved to: /content/drive/MyDrive/MAY19/final_bias_metrics_values.csv")
```

Normalize Values

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

bias_df = pd.read_csv("/content/drive/MyDrive/MAY19/final_bias_metrics_values.csv")
meta_df = pd.read_csv("/content/drive/MyDrive/MAY19/video_metadata_with_gender.csv")
bias_df = bias_df.merge(meta_df[["video_name", "gender"]], on="video_name", how="left")

# Select bias metric components
components = ["size_bias", "centering_bias", "screen_time_bias", "embedding_bias", "motion_bias"]

# === Standard Normalization (Z-score) ===
scaler = StandardScaler()
standardized = pd.DataFrame(
    scaler.fit_transform(bias_df[components]),
    columns=[f"{c}_std" for c in components]
)

# Append standardized columns
bias_df = pd.concat([bias_df, standardized], axis=1)

# === Aggregated Metrics ===
# 1. Mean of standardized scores (directional)
bias_df["combined_bias_std_mean"] = standardized.mean(axis=1)

# 2. Magnitude (sum of absolute standardized scores)
bias_df["combined_bias_std_magnitude"] = standardized.abs().sum(axis=1)

# 3. PCA-weighted score
pca = PCA(n_components=1)
bias_df["combined_bias_pca"] = pca.fit_transform(standardized)

# Rescale PCA score to [-1, +1]
bias_df["combined_bias_pca_scaled"] = MinMaxScaler(feature_range=(-1, 1)).fit_transform(
    bias_df[["combined_bias_pca"]]
)

output_path = "/content/drive/MyDrive/MAY19/final_bias_metrics_with_standardized.csv"
bias_df.to_csv(output_path, index=False)
print(f" Final standardized bias metrics saved to: {output_path}")
```
